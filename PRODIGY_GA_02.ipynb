## **Using Stable Diffusion**

```python

from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
!pip show torch
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe = pipe.to("cuda")

prompt = """ Pink Butterflies """

image = pipe(prompt).images[0]

print("[PROMPT]: ",prompt)
plt.imshow(image);
plt.axis('off');

```

## **Using DALL-E-mini**


```python
# Import the MinDalle module
from min_dalle import MinDalle

# Initialize the MinDalle model
model = MinDalle(is_mega=True, is_reusable=True, device="cuda")  # Use CUDA if available

# Define the text prompt and parameters
text = "Roses"
seed = 6
grid_size = 2

# Generate and display the image
image = model.generate_image(text, seed=seed, grid_size=grid_size)
image.show()  # This will display the generated image
```
